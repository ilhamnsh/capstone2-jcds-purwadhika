## 9. Membuat kolom revenue, potential_rev, loss_rev, date feature, grand total



### 9.1 Membuat kolom grand total baru
Terdapat anomali pada value dari kolom grand total.  
Grand total sendiri merupakan total harga yang dibayarkan oleh customer. Grand total memiliki rumus:  
grand_total = (price * qty_ordered) = discount amount  
  
Pada beberapa baris di dataset memiliki hasil anomali mengenai kolom grand_total. Seperti data di bawah :  
price = 360  
qty_ordered = 1  
discount_amount = 0  
grand total = 1270  

yang seharusnya memiliki grand total = 360  
Maka pembuatan kolom grand total baru dengan formulasi yang benar patut untuk dilakukan

```python
df[df['item_id'] == 211150.0]
```

```python
df = df.drop('grand_total',axis=1)
```

```python
df['grand_total'] = (df['price'] * df['qty_ordered']) - df['discount_amount']
```

### 9.2 Kolom Revenue
Pada kolom grand total tidak mencerminkan apakah total harga tersebut merupakan revenue / pendapatan bagi perusahaan ecommerce tersebut dikarenakan kondisi status transaksi yang terjadi. 
 
Pada kegiatan transaksi di E-commerce pakistan ini memiliki beberapa jenis status transaksi, yaitu :  
complete, canceled, recieved, order_refunded, refund, cod, paid, closed, payment_review, pending, processing, holded, fraud, pending_paypal, dan exchange

Sehingga dibutuhkan kolom revenue berdasarkan transaksi yang sudah selesai / completed

```python
df['revenue'] = df['price'] * df['qty_ordered'] - df['discount_amount']
df['revenue'] = df['revenue'][df['status'] == 'complete']
```

### 9.2 Membuat kolom potential_revenue
Potential revenue merupakan perkiraan revenue yang dapat didistribusikan sebagai revenue di masa depan. Potential revenue pada ecommerce ini merupakan revenue potensial yang bisa diperoleh oleh perusahaan dari proses transaksi, digambarkan oleh status transaksi, yang sedang terjadi. Potensial revenue tidak meliputi status transaksi yang sudah lengkap / completed. Pada kolom ini saya mengkategorikan potential revenue pada beberapa status transaksi yang terjadi. Diantaranya yaitu :  
recieved, order_refunded, refund, cod, paid, payment_review, pending, processing, holded, pending_paypal, dan exchange 

Maka rumus dari potential revenue adalah : ((price * qty_ordered) - discount_amount) dari transaksi yang disebutkan di atas

```python
df['pot_revenue'] = 0

valid_statuses = ['recieved', 'order_refunded', 'refund', 'cod', 'paid', 'payment_review', 'pending', 'processing', 'holded', 'pending_paypal', 'exchange']
mask = df['status'].isin(valid_statuses)
df.loc[mask, 'pot_revenue'] = df['price'] * df['qty_ordered'] - df['discount_amount']
```

### 9.3 Membuat kolom loss_revenue
Loss revenue merupakan pendapatan / revenue yang gagal didapatkan oleh perusahaan, dalam hal ini e-commerce pakistan. Loss revenue dalam hal ini dikategorikan berdasarkan status / kegiatan transaksi yang gagal. Diantaranya : canceled, fraud, closed

```python
df['loss_revenue'] = 0

loss_statuses = ['canceled', 'fraud', 'closed',]
mask1 = df['status'].isin(loss_statuses)
df.loc[mask1, 'loss_revenue'] = df['price'] * df['qty_ordered'] - df['discount_amount']
```

### 9.3 Membuat kolom Date Feature

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['dateofmonth'] = df['created_at'].dt.day
df['month'] = df['created_at'].dt.month
df['quarter'] = df['created_at'].dt.year.astype(str) + 'Q' + df['created_at'].dt.quarter.astype(str)
df['week'] = df['created_at'].dt.week
```

## 3. 1 RFM
RFM (Recency, Frequency, Monetary) merupakan sebuah metode untuk mensegmentasikan pelanggan kedalam beberapa klaster berdasarkan aktivitas nya dalam berbelanja. Perusahaan dapat mengembangkan analisa ini untuk mengetahui karakteristik dari customer/user nya. Metode ini berguna untuk menentukan strategi dan perencanaan untuk meningkatkan sales / revenue perusahaan  
  
RFM Analysis :
- Recency = Jumlah hari dari tanggal terakhir transaksi
- Frequency = Jumlah kuantitas order yang dilakukan 
- Monetary = Jumlah total harga yang dibayarkan customer di suatu transaksi

### 1. Data Preparation
- Mengubah datatipe 'Working Date' menjadi datatipe datetime
- Menentukan 'reference date' sebagai tanggal terakhir dalam data
- Filter 'most recent purchase date' dari setiap customer
- Kalkulasi recency dengan mensubtraksi 
- Menghapus kolom 'most recent purchase date'


#### 1.1 Recency

```python
df['working_date'] = pd.to_datetime(df['working_date'])
ref_date = df['working_date'].max()
df['Date'] = df['working_date'].apply(lambda x: x.date())
df['Most_Recent_Purchase'] = df.groupby('customer_id')['Date'].transform(max)
df['recency'] = (ref_date.date() - df.Most_Recent_Purchase).astype('timedelta64[D]')

```

```python
df = df.drop('Most_Recent_Purchase', axis=1)
```

#### 1.2 Frequency

```python
df['frequency'] = df.groupby('customer_id')['increment_id'].transform('nunique')
```

#### 1.2 Monetary

```python
df['monetary'] = df.groupby('customer_id')['grand_total'].transform(sum)
```

#### 1.5 Table RFM

```python
rfm = df[['customer_id', 'recency', 'frequency', 'monetary']].drop_duplicates()
rfm
```

```python
rfm.describe().T
```

### 2. Membuat RFM
#### 2. 1 Menentukan Kuartil RFM

```python
quantiles = rfm.quantile(q=[0.25,0.5,0.75])
```

```python
quantiles
```

Ubah quantiles ke dictionary untuk memudahkan

```python
quantiles = quantiles.to_dict()
quantiles
```

#### 2.2 Membuat RFM Segment
Membuat 2 jenis RFM segment karena recency yang tinggi tidak baik, sementara frequency dan monetary yang tinggi adalah baik


```python
rfm_segment = rfm
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def RClass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

#### 2.3 RFM Scores
Range skor :
- 1 = Paling tinggi
- 2 = Tinggi
- 3 = Normal
- 4 = Rendah
- 5 = Paling rendah

```python
rfm_segment['R_Quartile'] = rfm_segment['recency'].apply(RClass, args=('recency',quantiles,))
rfm_segment['F_Quartile'] = rfm_segment['frequency'].apply(FMClass, args=('frequency',quantiles,))
rfm_segment['M_Quartile'] = rfm_segment['monetary'].apply(FMClass, args=('monetary',quantiles,))

```

```python
rfm_segment['rfm_score'] = rfm_segment['R_Quartile'] + rfm_segment['F_Quartile'] + rfm_segment['M_Quartile']
```

```python
rfm_segment
```

#### 2.4 Customer Segmentation : RFM Score

```python
rfm_segment.loc[(rfm_segment['rfm_score'] == 12.0), 'cluster'] = 'lost_customers'
rfm_segment.loc[(rfm_segment['rfm_score'] == 11.0), 'cluster'] = 'hibernating'
rfm_segment.loc[(rfm_segment['rfm_score'] >= 9.0) &  (rfm_segment['rfm_score'] <= 10.0),'cluster'] = 'at risk'
rfm_segment.loc[(rfm_segment['rfm_score'] >= 7.0) &  (rfm_segment['rfm_score'] <= 8.0),'cluster'] = 'promising customers'
rfm_segment.loc[(rfm_segment['rfm_score'] >= 4.0) &  (rfm_segment['rfm_score'] <= 6.0),'cluster'] = 'loyal_customers'
rfm_segment.loc[(rfm_segment['rfm_score'] == 3.0), 'cluster'] = 'champions'
```
